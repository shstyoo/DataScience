```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)

avg_stats_all = pd.read_json('results_core_validated.json', convert_axes=True)
```

```python
df = avg_stats_all.iloc[[14]]
fig, ax = plt.subplots()
hero_list_x = []
winrate_y = []
for item in df:
    hero_list_x.append(item)
    winrate_y.append(df[item]['WINRATE'])

    
n_group = 22
x = np.arange(22)
y = winrate_y
labels = hero_list_x
plt.plot(x, y, 'ro')
plt.xticks(x, labels, rotation='vertical')
plt.margins(0.2)
plt.show()
```
